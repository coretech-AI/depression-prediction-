```python
# Check the number of missing values per column
missing_values = df.isnull().sum()

# Check the percentage of missing values per column
missing_percentage = (missing_values / len(df)) * 100

# Display missing value statistics
print(pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage}))
```

```python
df["Job Satisfaction"]
```

```python
df["Job/study Satisfaction"] = df["Job Satisfaction"].combine_first(df["Study Satisfaction"])
```

```python
df["Job/study Satisfaction"]
```

```python
df.drop(columns=["Study Satisfaction", "Job Satisfaction"], inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=["id", "Name"], inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=["Profession", "CGPA"], inplace=True)
```

```python
df.head()
```

```python
df["Have you ever had suicidal thoughts ?"] = df["Have you ever had suicidal thoughts ?"].map({"Yes": 1, "No": 0})
df["Family History of Mental Illness"] = df["Family History of Mental Illness"].map({"Yes": 1, "No": 0})
```

```python
df.head()
```

```python
sleep_mapping = {"Less than 5 hours": 1, "5-6 hours": 2, "More than 8 hours": 3}
df["Sleep Duration"] = df["Sleep Duration"].map(sleep_mapping)

diet_mapping = {"Unhealthy": 1, "Moderate": 2, "Healthy": 3}
df["Dietary Habits"] = df["Dietary Habits"].map(diet_mapping)
```

```python
df.head()
```

```python
df["Academic/Work_Pressure"] = df["Academic Pressure"].combine_first(df["Work Pressure"])
```

```python
df.head()
```

```python
df.drop(columns=["Academic Pressure", "Work Pressure"], inplace=True)
df.head()
```

```python
working_map={"Working Professional":1, "Student":0}
df["Working Professional or Student"]=df["Working Professional or Student"].map(working_map)
```

```python
df.head()
```

```python
uniq=unique_values = df['City'].unique()
print(uniq)
```

```python
import numpy as np

# Given list
city_list = np.array(uniq)  # np.nan included

# Remove NaN values and create dictionary
city_dict = {city: index for index, city in enumerate(city_list) if isinstance(city, str)}

# Print the dictionary
print(city_dict)
df["City"]=df["City"].map(city_dict)
```

```python
df.head()
```

```python
uniq=unique_values = df['Degree'].unique()
print(uniq)
```

```python
degree_list = np.array(uniq)  # np.nan included

# Remove NaN values and create dictionary
degree_dict = {degree: index for index, degree in enumerate(degree_list) if isinstance(degree, str)}

# Print the dictionary
print(degree_dict)
df["Degree"]=df["Degree"].map(degree_dict)
```

```python
df.head()
```

```python
gender_map={"Female":0, "Male":1}
df["Gender"]=df["Gender"].map(gender_map)
```

```python
df.head()
```

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
